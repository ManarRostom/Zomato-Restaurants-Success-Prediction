<span style="color:#2929a3;font-size:20px;">Create " merge_rows_of_same_restaurant " Function</span>

#### This function is designed to eliminate duplicated rows that exist but have distinct values in the 'listed_in(type)' column. The unique 'listed_in(type)' values are then combined into a single string separated by commas

```python
def merge_rows_of_same_restaurant(row):
    ## List of "listed_in(type)" of specific restaurant
    list_of_types = list(set(df[df['base_url'] == row['base_url']]['listed_in(type)'].tolist()))
    
    ## indicies of duplicated rows of specific restaurant
    same_rows_indices = df[df['base_url'] == row['base_url']].index
    
    ## Access duplicated rows and put in "listed_in(type)" column the combined string that is separated by commas 
    df.loc[same_rows_indices,'listed_in(type)'] = ",".join(list_of_types)
    
    ## Drop Rows that has the same name , rate and address
    df.loc[same_rows_indices] = df.loc[same_rows_indices].drop_duplicates(subset=['name','rate','address'],keep='first')
    df.dropna(how='all',inplace=True)
    
    ## Reset index
    df.reset_index(drop=True, inplace=True)
```

```python
## Apply merge_rows_of_same_restaurant Function
r = df.apply(lambda row:merge_rows_of_same_restaurant(row), axis=1)
```
